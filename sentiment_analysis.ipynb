```python
import pandas as pd

file_path = "/content/drive/My Drive/IMDB-movie-review-sentiment-analysis/data/archive/IMDB Dataset.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
import re
from typing import Dict

def clean_html_text(data: Dict) -> Dict:
    """
    Purpose:
    Removes HTML tags from the dataset.

    Args:
    data (Dict): Dictionary representing one row of the dataset,
    expected to have a 'review' key.

    Returns:
    The same dictionary with the 'review' field cleaned of HTML tags.

    """

    html_pattern = re.compile('<.*?>')

    # Replacing HTML tags with a space, and updating it.
    if 'review' in data and isinstance(data['review'], str):
        data['review'] = re.sub(html_pattern, ' ', data['review'])

    return data
```

```python
from datasets import Dataset
raw_dataset = Dataset.from_pandas(df)
cleaned_dataset = raw_dataset.map(clean_html_text)
```
